## Mutations + Selection (Children plus Parent)

```python
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt

def flip(state, k):
    fl = np.random.choice(range(100), k, replace = False)
    if state[fl].any() == 0:
        state[fl] = 1
    else:
        state[fl] = 0
    return state

def families_per_gen(state, children):    
    fam = np.vstack((np.asarray(children), state))
    return fam

def fam_sum(fam):
    fam_sum = np.asarray([sum(i) for i in fam])
    return fam_sum

def max_member_index(fam_sum):
    max_m = np.argmax(fam_sum)    
    return max_m

def total_array(big_loop, generations, C, flips):
    for b in range(big_loop):
        state = np.zeros(100) 
        for gen in generations:
                all_children = [flip(state.copy(), flips) for i in range(C)]
                fam = families_per_gen(state, all_children)
                f_sum = fam_sum(fam)
                mx = max_member_index(f_sum)
                state = fam[mx]
                generation_list.append(state.sum())
    generation_ar = np.asarray(generation_list)

    big_list = generation_ar.reshape(-1, 200)
    return big_list

generation_list = []
big_loop = 1000
generations = range(200)
```

## Children = 1

```python
big_array1 = total_array(big_loop, generations, C = 1, flips = 1)
```

```python

list_per_generation = [[item[index] for item in big_array1] for index in range(200)]
mean_per_generation = [np.mean(gen_x) for gen_x in list_per_generation]
mean_array = np.asarray(mean_per_generation)
```

```python
mean_dif = np.diff(mean_array)
mean_dif.shape
```

```python
y = np.arange(200)

plt.ylabel("Generation Component Sum")
plt.xlabel("Generation")
plt.title("Component Sum Vs. Generation")

for i in big_array:
    plt.plot(y,i)
plt.plot(y,mean_per_generation, linewidth=4, color = 'C2', label = 'Mean')
plt.legend()
plt.show()
```

```python
y_dif = np.arange(199)
plt.plot(y_dif, mean_dif, label = "Real") 
E_imp1 = 1 - (np.arange(2, 101) / 100) ** 1
# plt.plot(np.arange(2, 101), E_imp1, c='b', alpha=0.8, label='Theoretical')
plt.legend()
plt.ylabel("Difference")
plt.xlabel("Generation")
plt.title("Mean Difference per Generation")
plt.show()
```
