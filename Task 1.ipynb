## Mutations only â€“ Ignore the Parent

```python
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt



generation_list = []
sum_list = []
big_loop = 100
big_list = []
for b in range(big_loop):
    init_state = np.zeros(100)
    for l in range(200):

        i = random.choice(range(100))

        if init_state[i] == 0:
            init_state[i] = 1
        else:
            init_state[i] = 0

        generation_list.append(init_state.copy())

        generation_ar = np.asarray(generation_list)

for gen in generation_ar:
    sum_list.append(sum(gen.copy()))

    
sum_arr = np.asarray(sum_list)
```

## Composed Array

```python
big_list = sum_arr.reshape(-1, 200)
```

## Mean per Generation

```python

list_per_generation = [[item[index] for item in big_list] for index in range(200)]
mean_per_generation = [np.mean(gen_x) for gen_x in list_per_generation]
mean_array = np.asarray(mean_per_generation)
```

## Difference between new generation mean and old generation mean

```python
mean_dif = np.diff(mean_array)
mean_dif.shape
```

```python
y = np.arange(200)

plt.ylabel("Generation Component Sum")
plt.xlabel("Generation")
plt.title("Component Sum Vs. Generation")

for i in big_list:
    plt.plot(y,i)
plt.plot(y,mean_per_generation, linewidth=4, color = 'C2', label = 'Mean')
plt.legend()
plt.show()
```
